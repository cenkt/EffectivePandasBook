```python
t = pd.Series(range(1, 60))
cosine = np.cos(2 * np.pi * t / 10)
line = -2 + 0.05 * t
xl = cosine + line
xs = xl.rolling(5, center=True).mean()
ax = xl.plot()
xs.plot(ax=ax)
```

### Using Moving Average Smoothing for Estimating Trend in Data with Fixed Cycle Lengths


```python
air_passengers = pd.read_csv("AirPassengers.csv", parse_dates=["Date"], index_col="Date")
air_passengers.index = air_passengers.index + pd.DateOffset(years=-100)
air_passengers.index.name = None
air_passengers.plot()
```

```python
ax = air_passengers.rolling(12).mean().plot()
air_passengers.rolling(15).mean().plot(ax=ax)
ax.legend(["12 Smooting", "15 Smooting"])
```

```python
unemp = pd.read_csv("tx.unemp.unadj.csv")
dti = pd.date_range("2000", "2020", freq="M")
unemp.index = dti
unemp_z = unemp["2016-01-31":]
```

```python
# ax=unemp.plot()
# unemp.rolling(12).mean().plot(ax=ax)
ax = unemp_z.plot()
unemp_z.rolling(12).mean().plot(ax=ax)
```

<span style="color:red">_Key Point: A moving average smoother of order k applied to cyclic data with a fixed cycle length, k, will remove the cyclic behavior and retain the trend even when the trend follows a wandering path
as in the Texas Unemployment data._</span>


### 2-Decomposing Seasonal Data


#### a-Estimate the Trend Component (Additive Decompositions)


```python
air_passengers.plot()
```

```python
air_passengers.plot(logy=True)
```

```python
air_passengers["LogNumPas"] = np.log(air_passengers["NumPassengers"])
ax = air_passengers["LogNumPas"].plot()
air_passengers["LogNumPas"].rolling(
    12,
    center=True,
).mean().plot(ax=ax)
```

#### b-Remove the Trend Component from the Data by Subtructing the Trend


```python
no_trend = air_passengers["LogNumPas"] - air_passengers["LogNumPas"].rolling(12, center=True).mean()
no_trend.plot()
```

#### c-Calculate an “Average” Within-year Seasonal Component


```python
monthly_average = no_trend.groupby([no_trend.index.month]).transform("mean")
monthly_average.plot(ylim=(-0.6, 0.6))
```

#### d-Find the Remaining Noise Component


```python
(no_trend - monthly_average).plot(ylim=(-0.6, 0.6))
```

```python
figure, axes = plt.subplots(2, 2)
plt.subplots_adjust(hspace=0.4)
air_passengers["LogNumPas"].plot(ax=axes[0, 0], title="log AirPassengers data")
air_passengers["LogNumPas"].rolling(
    12,
    center=True,
).mean().plot(ax=axes[0, 1], title="estimated trend component")
monthly_average.plot(ax=axes[1, 0], ylim=(-0.6, 0.6), title="estimated seasonal component")
(no_trend - monthly_average).plot(ax=axes[1, 1], ylim=(-0.6, 0.6), title="noise component")
```

#### abcd-Estimate the Trend Component (Multiplicative Decompositions)


```python
air_passengers["NumPassengers"].rolling(
    12,
    center=True,
).mean().plot()
```

```python
no_trend2 = air_passengers["NumPassengers"] / air_passengers["NumPassengers"].rolling(12, center=True).mean()
no_trend2.plot()
```

```python
monthly_average2 = no_trend2.groupby([no_trend2.index.month]).transform("mean")
# monthly_average2
```

```python
figure, axes = plt.subplots(2, 2)
plt.subplots_adjust(hspace=0.4)
air_passengers["NumPassengers"].plot(ax=axes[0, 0], title="Air Passengers data")
air_passengers["NumPassengers"].rolling(
    12,
    center=True,
).mean().plot(ax=axes[0, 1], title="estimated trend component")
monthly_average2.plot(ax=axes[1, 0], ylim=(0.4, 1.6), title="estimated seasonal component")
(no_trend2 / monthly_average2).plot(ax=axes[1, 1], ylim=(0.4, 1.6), title="noise component")
```

#### Texas Unemoloyment


```python
dates = pd.date_range(start="01-2000", end="01-2020", freq="M")
tex_unp = pd.read_csv("tx.unemp.unadj.csv")
tex_unp.index = dates
tex_unp.plot()
```

```python
ternd_tex_unp = tex_unp.rolling(12, center=True).mean()
ternd_tex_unp.plot()
```

```python
(tex_unp - ternd_tex_unp).plot()
```

```python
season_tex_enp = (tex_unp - ternd_tex_unp).groupby(tex_unp.index.month).transform("mean")
season_tex_enp.plot()
```

```python
((tex_unp - ternd_tex_unp) - season_tex_enp).plot()
```

```python
figure, axes = plt.subplots(2, 2)
plt.subplots_adjust(hspace=0.4)
tex_unp.plot(ax=axes[0, 0], title="Air Passengers data")
tex_unp.rolling(
    12,
    center=True,
).mean().plot(ax=axes[0, 1], title="estimated trend component")
season_tex_enp.plot(ax=axes[1, 0], ylim=(-1.5, 1.5), title="estimated seasonal component")
((tex_unp - ternd_tex_unp) - season_tex_enp).plot(ax=axes[1, 1], ylim=(-1.5, 1.5), title="noise component")
```

```python
dates = pd.date_range(start="01-1990", end="01-2021", freq="M")
wtc = pd.read_csv("wtcrude2020.csv")
wtc.index = dates
wtc36 = wtc.loc["01-2018":"12-2020"][:]
wtc36
```

```python
fit1 = SimpleExpSmoothing(wtc36.x, initialization_method="heuristic").fit(
    smoothing_level=0.4, optimized=False
)
fcast1 = fit1.forecast(3).rename(r"$\alpha=0.2$")
fcast1
```

```python
ax = wtc36.plot()
fit1.fittedvalues.plot(ax=ax)
fcast1.plot(ax=ax)
```
